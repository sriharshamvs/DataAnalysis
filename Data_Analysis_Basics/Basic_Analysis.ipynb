```python
import pandas as pd
```

```python
records = [
    {'name': 'Luke Skywalker', 'age': 60, 'power-level':95, 'type': 'Jedi'},
    {'name': 'Han Solo', 'age': 65},
    {'name': 'Leia', 'age': 65, 'power-level': 50},
    {'name': 'Darth Wadar', 'age': 60, 'power-level': 95, 'type': 'Jedi'},
    {'name': 'Yoda', 'age': 500, 'power-level': 100, 'type': 'Jedi'},
    {'name': 'Rey', 'age': 19, 'power-level': 80, 'type': 'Jedi'},
    {'name': 'Kylo Ren', 'age': 20, 'power-level': 82, 'type': 'Jedi'}
]
```

```python
SW_char = pd.DataFrame(records)
SW_char
```

```python
columns = ['Age', 'Name', 'PowerLevel', 'Type']
SW_char.columns = columns
order = ['Name', 'Age', 'PowerLevel', 'Type']
SW_char = SW_char[order]
SW_char
```

```python
SW_char['PowerLevel'] = SW_char['PowerLevel'].fillna(50)
SW_char
```
